```python
import pandas as pd
data1 = pd.read_csv('For_sampling_edit (0816).csv')
data1.head()
```

```python
data = data1.drop(['living_year_B', 'living_year_B2'], axis = 1)
```

```python
data
```

```python
data.shape
```

```python
# y 불균형 정도 보기 위해, 더미별 개수 확인
data.discrimination.value_counts()
```

```python
# *** X값(독립변수) y(종속변수)값 설정하기
X = data.iloc[:,:-1]
y = data.discrimination
X.head()
```

```python
pip install imblearn
```

```python
# SMOTE 분석 세팅
from imblearn.over_sampling import SMOTE
smote = SMOTE()
```

```python
# SMOTE값 도출하기 
X_smote, y_smote = smote.fit_resample(X.astype('float'),y)
```

```python
# SMOTE값 이전 이후 보기
from collections import Counter
print("Before SMOTE :" , Counter(y))
print("After SMOTE :" , Counter(y_smote))
```

```python
# 독립변수 보기
X_smote
```

```python
# y값 보기
y_smote
```

```python
# y값 유형별 개수 보기
y_smote.value_counts()
```

```python
SMOTE_table = pd.concat([X_smote, y_smote], axis = 1)
```

```python
SMOTE_table
```

Under Sampling 해보기

```python
from imblearn.under_sampling import RandomUnderSampler 
rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(X, y)
print("Before SMOTE :" , Counter(y))
print("After SMOTE :" , Counter(y_res))
```

```python
X_res
```

```python
from imblearn.under_sampling import CondensedNearestNeighbour

undersample = CondensedNearestNeighbour(n_neighbors = 10)
X_Cnn, y_Cnn = undersample.fit_resample(X, y)
print("Before CNN :" , Counter(y))
print("After CNN :" , Counter(y_Cnn))
```

```python
Cnn_table = pd.concat([X_Cnn, y_Cnn], axis = 1)
```

```python
Cnn_table.to_excel(
    excel_writer = 'Cnn_result.xlsx',
    sheet_name = 'Sheet1',
    index = True,       # 0부터 시작하는 자연수 인덱스는 의미가 없음.
    # columns = ['col1, 'col2', 'col3'],
    encoding = 'utf-8',
    na_rep = '',      # 결측값을 ''으로 채우기
    inf_rep = ''     # 무한값을 ''으로 채우기
)     # 해당 파일이 열려있으면 안됨.
```

```python
from imblearn.under_sampling import OneSidedSelection

oss = OneSidedSelection(n_neighbors = 200, n_seeds_S = 400)
X_Oss, y_Oss = oss.fit_resample(X, y)
print("Before OSS :" , Counter(y))
print("After OSS :" , Counter(y_Oss))
```

```python
from imblearn.under_sampling import NearMiss

# define the undersampling method
undersample2=NearMiss(version=1, n_neighbors_ver3=1)
# transform the dataset
X_Under, y_Under = undersample2.fit_resample(X, y)
print("Before NM :" , Counter(y))
print("After NM :" , Counter(X_Under))
```

```python
from imblearn.under_sampling import TomekLinks

tl = TomekLinks()
X_tl, y_tl = tl.fit_resample(X, y)
print("Before TL :" , Counter(y))
print("After TL :" , Counter(y_tl))
```
